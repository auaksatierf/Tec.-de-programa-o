## Aplicação com Listas, Tuplas é Dicionarios.

```python
alunos_db = [
    {"matricula": "676", "nome": "Raimundo Neto", "curso": "Engenharia de Software"},
    {"matricula": "1180", "nome": "Gustavo Barbosa", "curso": "Psicologia"},
    {"matricula": "4456", "nome": "Saymon Dias", "curso": "Analise é Densenvolvimento de Sistemas"}
]

print("1. Sistema Simplificado de Cadastro de Aluno")
print("_" * 100)

#1_Adicionar um novo aluno(CREATE)
print("-> Adiciona novo aluno...")
novo_aluno = {"matricula": "782", "nome": "Anthoy Leão", "curso": "Rede de Computadores"}
print(f"Aluno {novo_aluno['nome']} adicionado com sucesso.")
matricula_busca = "676"
aluno_encontrado = next((aluno for aluno in alunos_db if aluno["matricula"] == matricula_busca), None)

if aluno_encontrado:
  print(f"\nAluno encontrado (Matricula: {matricula_busca}):" )
else:
  print(f"\nAluno com matricula {matricula_busca} não encontrado.")

matricula_update = "676"
for aluno in alunos_db:
  if aluno["matricula"] == matricula_update:
      aluno["curso"] = "Segurança da Infomatica"
      print(f"\nCurso do aluno {aluno['nome']} atulizando para: {aluno['curso']}")
      break
print("_" * 100)
```
